READING DATA USING PANDAS

We use pandas read_csv function to read  csv file in python and pandas DataFrame method to convert file into the data frame.

```python
df = pd.DataFrame(pd.read_csv('/content/train.csv'))
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
 # df.isnull().sum() returns a panda series with column names as the label index
 # and totsl count of null values in the column as its value 
 # And we are storing only those colunms which have more than 35% of the data missing.

x = df.isnull().sum()
drop_col = x[x>(35/100 * df.shape[0])]
drop_col
```

```python
drop_col.index
```

```python
df.drop(drop_col.index, axis=1, inplace=True)
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace = True)
df.isnull().sum()
```

```python
df['Embarked'].describe()
```

```python
df['Embarked'].fillna('S',inplace=True)
```

```python
df.isnull().sum()       
```

```python
df.corr()
```

```python
df['FamilySize']=df['SibSp']+df['Parch']
df.drop(['SibSp','Parch'],axis=1, inplace=True)
df.corr()
```

```python
df['Alone']= [0 if df['FamilySize'][i]>0 else 1 for i in df.index]
df.head()
```

```python
df.groupby(['Alone'])['Survived'].mean()
```

```python
df[['Alone','Fare']].corr()
```

```python
df['Sex'] = [0 if df['Sex'][i]=='male' else 1 for i in df.index]
df.groupby(['Sex'])['Survived'].mean()
```

```python
df.groupby(['Embarked'])['Survived'].mean()
```
